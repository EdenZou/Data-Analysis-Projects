```python
from selenium import webdriver
import time
import csv
import pandas as pd

links = []
data = pd.read_csv('../data/Bilibili_outside_funny.csv',usecols=[0])
# data = pd.read_csv('../data/Bilibili_outside_food.csv',usecols=[0])
# data = pd.read_csv('../data/Bilibili_outside_makeup.csv',usecols=[0])
# data = pd.read_csv('../data/Bilibili_outside_funny.csv',usecols=[0])

for row in range(5000):
    row =data.loc[row]['link']
    links.append(row)
    
# print(links)

def isElementExist(xpath):
    try:
        driver.find_element_by_xpath(xpath)
        return True
    except:
        return False

# x = ['https://www.bilibili.com/video/av80623902?','https://www.bilibili.com/video/av80683504?']
Data1 = []
driver = webdriver.Chrome(r"//Users/binxiwu/Documents/forSelenium/chromedriver")
i=1
for url in links:
    driver.get(url)
    time.sleep(8)


    # get the publish time 
    publish = driver.find_element_by_xpath("//*[@id='viewbox_report']/div[1]/span[2]").text
    
    # get the numbers of like
    like = driver.find_element_by_xpath("//*[@id='arc_toolbar_report']/div[1]/span[1]").text
    
    # get the number of medal
    toujinbi = driver.find_element_by_xpath("//*[@id='arc_toolbar_report']/div[1]/span[2]").text
    
    # get the number of collections
    collect = driver.find_element_by_xpath("//*[@id='arc_toolbar_report']/div[1]/span[3]").text
    
    # get the number of reposts
    repost = driver.find_element_by_xpath("//*[@id='arc_toolbar_report']/div[1]/span[4]").text
    
    # if there are more than one autho, we ignore the authors
    if isElementExist('//*[@id="v_upinfo"]/div[3]/div/span/span'):
        follower = driver.find_element_by_xpath('//*[@id="v_upinfo"]/div[3]/div/span/span').text
    else:
        follower = 'none'

    Data1.append([url, publish, like, toujinbi, collect, repost, follower])
    print(i)
    i=i+1

outfile = open('final_bilibili_perfect_fitness.csv', 'w', newline='')
writer = csv.writer(outfile)
writer.writerow(['url', 'publish', 'like', 'toujinbi', 'collect', 'repost', 'follower'])
writer.writerows(Data1)
outfile.close()
driver.close()
```
